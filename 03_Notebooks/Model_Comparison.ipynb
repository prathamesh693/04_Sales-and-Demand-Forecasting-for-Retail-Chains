```python
# Required Libraries
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import root_mean_squared_error, mean_squared_error
```

```python
# Load prediction data
df = pd.read_csv("R:/Projects/1_Data_Science & ML_Projects/04_Sales and Demand Forecasting for Retail Chains or Predictive Sales/02_Data/model_predictions.csv", parse_dates=["Date"])

store_id = 1
df = df[df["Store"] == store_id]
```

```python
# Actual and predictions
actual = df["Weekly_Sales"]
arima = df["ARIMA_Prediction"]
prophet = df["Prophet_Prediction"]
lstm = df["LSTM_Prediction"]
```

```python
# Plot all
plt.figure(figsize=(14,6))
plt.plot(df["Date"], actual, label="Actual", color="black")
plt.plot(df["Date"], arima, label="ARIMA", linestyle="--")
plt.plot(df["Date"], prophet, label="Prophet", linestyle=":")
plt.plot(df["Date"], lstm, label="LSTM", linestyle="-.")

plt.title("Model Comparison: Actual vs Predictions")
plt.xlabel("Date")
plt.ylabel("Weekly Sales")
plt.legend()
plt.show()
```
